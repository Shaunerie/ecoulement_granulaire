## I. Présentation générale
---
<a id="Ia"></a>
### I.a) Résumé du projet :

<p style="text-align: justify;"> <b>Les matériaux granulaires représentent un enjeu majeur de notre société</b>; ils sont présents dans tout les domaines sous différentes formes, du sable à la roche en passant par le blé. Comprendre leurs comportements dans les milieux qui les entourent est donc essentiel pour mettre des avancées dans de nombreux secteurs (géologie, industries minières/pharmaceutique/chimique...) mais également pour optimiser leur transport et leur stockage. </p>

<p style="text-align: justify;"> Dans notre projet, nous nous sommes concentrés sur le secteur de l'agroalimentaire avec les silos à grains. L'objectif principal du projet est donc de créer une simulation, en python, permettant de vérifier numériquement une formule analytique du débit granulaire à la sortie d'un silo à grain. </p>

<a id="Ib"></a>
### I.b) Brève introduction à la modélisation numérique :
<p style="text-align: justify;"> Afin réaliser cette modélisation, nous utilisons une méthode de discrétisation appelée <b>Discrete Element Method (DEM)</b>. Cette méthode permet de simuler les interactions entre les grains et les parois du silo en considérant chaque grain comme une entité discrète. </p>

<p style="text-align: justify;"> Chaque grain, possède sa propre taille et masse. A chaque instant on associe une position, une vitesse et une accélération à tous les grains de notre simulation. </p>


<p style="text-align: justify;"> Pour réaliser la simulation il nous faut aussi discrétiser le temps. On définit une durée pour la simulation ainsi qu'un pas de temps déterminant l'intervalle d'actualisation des caractéristique de nos grains. Autrement dit, tous les pas de temps, nous mettons à jour les intéractions sur les grains afin de calculer leur nouvelle position, vitesse et accélération. </p>

## II. Démarche scientifique 
---

<p style="text-align: justify;"> Pour élaborer notre approche scientifique et les méthodes numériques à utiliser, nous avons principalement utilisé le cours et les ressources fournies par notre tuteur. </p>

<a id="IIa"></a>
### II.a) Contacts et paramètres physiques :

<p style="text-align: justify;"> Parmi ces méthodes, on retrouve la modélisation des contacts entre les différents éléments de notre simulation. </p>

<p style="text-align: justify;"> En vue de modéliser un contact entre deux entités, on introduit des ressorts ainsi qu'un amortisseur entre les deux grains. Cette approche permet de mettre en oeuvre les efforts repulsif et attractifs qui découlent d'un contact entre deux objets. Ci-dessous un schéma descriptif: </p>
<a id="figure1"></a>
<figure style="text-align: center;">
    <img src=ressort.png align="center" width=600>
    <figcaption style="border: 1px solid black; padding: 10px; display: inline-block;"><b>Fig.1 - Représentation schématique du contact entre deux grains</b></figcaption>
</figure>

<p style="text-align: justify;"> Les valeurs $ \delta_n $ et $\delta_t $ représentent respectivement l'allongement normal et tangentiel. Ils représentent l'interpénétration des deux grains. Plus les deux objets s'interpénetrent plus les efforts repulsifs sont importants. Nous pouvons faire l'analogie avec un ressort: plus on compresse un ressort plus la réaction du ressort est importante </p>

<p style="text-align: justify;"> Les paramètres $k_n, \gamma_n, \mu $ sont des constantes définissant respectivement la raideur, le coefficient d'amortissement et le coefficient de frottement. </p>

<p style="text-align: justify;"> Ces paramètres permettent de caractériser et de modifier manuellement les intéractions qui ont lieu. En augmentant la raideur, par exemple, nous amplifions la réaction du ressort proportionnellement à son allongement. </p>
    
<a id="ref_relation"></a>
<p style="text-align: justify;"> <b>Le choix de toutes ces valeurs est crucial pour le bon fonctionnement de la simulation</b>. En effet, si les paramètres sont incohérents et/ou incohérents entre eux, bien que le code soit juste, la simulation sera fausse. Le choix de ses valeurs est arbitraire et doit être changé manuellement pour s'adapter à chaque modélisation. Cependant, il existe des formules usuelles reliant ces grandeurs entre-elles fournissant des résultats satisfaisant dans la plupart des temps. </p>

<p style="text-align: justify;"> Prenons l'exemple de l'effort normal au contact entre deux grains : </p>
    
<a id="figure2"></a>
<figure style="text-align: center;">
    <img src="equation_ressort.png" alt="Résultante de l'effort normal" width="600">
    <figcaption style="border: 1px solid black; padding: 10px; display: inline-block;"><b>Fig.2 - Résultante de l'effort normal</b></figcaption>
</figure>



<p style="text-align: justify;"> Le terme $k_n \cdot \delta_n$ représente l'effort répulsif lorsque deux grains s'interpénètrent, c'est le terme analogique à la résultante d'un ressort. </p>

<p style="text-align: justify;"> Le terme $\gamma_n \cdot \frac{d\delta_n}{dt}$ représente l'amortissement . Cette valeur permet d'amortir les contacts trop intenses elle est répulsive au début de l'impact et attractive à sa fin. </p>

<a id="IIb"></a>
### II.b) Dynamique et schéma d'intégration temporel :

<p style="text-align: justify;"> En dehors des équations modélisant les efforts, si nous voulons modéliser la dynamique de nos particules, nous devons faire appel aux Principe Fondamental de la Dynamique. Le PFD permet de relier l'accélération d'un système aux efforts extérieurs qu'il subit. A partir du PFD de Newton, nous pouvons quantifier l'accélération d'un système via le bilan des actions extérieurs qui s'appliquent sur celui-ci. </p>

<p style="text-align: justify;"> Néamoins, dans le but de prévoir l'évolution de la vitesse et de la position d'un grain, il est nécessaire d'intégrer cette équation fondamentale. C'est pourquoi, afin d'intégrer temporellement ces équations nous devons faire appel à un "<b>schéma d'intégration temporel</b>". L'intégration temporelle est l'ensemble des techniques qui permettent de résoudre numériquement des équations différentiel les ordinaires, qui sont des équations décrivant le comportement d'une variable au fil du temps. Ici le schéma est utilisé pour le PFD et les équations qui en dérivent. </p>

<p style="text-align: justify;"> Une fois que nous avons choisi et appliqué un schéma d'intégration, nous pouvons prédire l'évolution de la position et de la vitesse de chaque particule au fil du temps, en fonction des forces externes qu'elle subit. </p>

<p style="text-align: justify;"> <b>Le schéma d'intégration temporel utilisé dans notre projet est celui de Verlet</b> que nous pouvons décrire étape par étape comme suit: </p>

</figure>

Les équations mentionnées telles que 'Équations (4.18)' sont les équations temporelles qui rendent comptent de l'application du schéma d'intégration temporel au PFD sur les grains de notre simulation: </p>

<p style="text-align: justify;"> Le terme $\Delta t$ est le pas de temps de notre simulation. <b>C'est probablement le paramètre le plus important de la simulation</b>. En effet, plus un pas de temps est petit, plus le caractère discret du temps se rapprochera du caratère continue rélle. U  de temps faible inclue une simulation plus longue car impliquant un plus grand nombre de calcul pour une simulation à durée fixée. De plus, und nombre de calcul, résulte souvent en une occpuation de mémoire(numérique) plus importante. Ce paramètre est à la fois une appréciation de la qualité de la simulation car central au cœur des calculs réalisé <b>Par exemple</b>, généralement, l'une des première erreurs que fait une personne est d'attribuer une trop grande <a href="#figure1"> raideur </a> pour un pas de temps pas suffisamment faible. La trop grande raideur conduit à une réaction dynamique très forte qu'il est nécessaire de suivre rapidement. Sans un pas de temps faible, il est impossible de suivre la forte réponse induite par l'importante raideur choisie. </p>

### III.a) Simplifications :
- <p style="text-align: justify;"> Notre modèle ne considère pas la rotation des grains. Autrement dit, <b>nos grains ne 'roulent' pas, ils glissent</b>. Notre modèle se limite donc à deux degrés de liberté (l'abscisse et l'ordonné) soit une <b>modélisation en 2D</b>. Voici une liste des différents phénomènes physiques implémentés :</p>

    + Contact grain-grain, grain-paroi, grain-bac:
        + Résultante normale.
        + Amortissement.
        + Résultante tangentielle.
        + Glissement.
    + Autres:
        + Efforts à distance (gravite due au poids des grains)
        + Frottement de l'air (trainée aéronotique)

Toutefois,</b> les rayons des grains varient de <a id="variation_rayon"></a>$\pm$20% par rapport à une valeur de référence. La masse associé aux grains et calculée via la masse volumique des grains. </p>

En effet, lorsqu'on réalise des simulations numériques, le rendu final prend généralement énormément de temps à être finalisé. C'est pourquoi, pour gagner du temps sur nos calculs, on évite généralement de calculer toutes les variables entre tous les grains et toutes les parois à chaque instant de la simulation. A la place, on implémente une liste de contact qui attribue à chaque grain les contacts qui réalise avec son environnement(autres grains ou parois). Pour réaliser cette liste de contact, on utilise alors une liste de voisinage propre à chaque grain qui rend compte de l'environnement autour de l'objet. Pour notre simulateur <b>nous avons choisi un rayon de référence de 5 $mm$, ainsi qu'une masse volumique de 770 $kg / m^3$</b>. Autrement dit, pour tous les grains leur rayon sera compris tel que: $ 4 < r < 6,$ en $mm$ et la masse correspondante: $m=\frac{4}{3} \pi r^3 \rho \space(kg)$, avec $p$ la masse volumique et $r$ le rayon. Ces deux valeurs ne sont pas anodines elles impactent grandement la simulation et il est important de les choisir avec précaution. C'est pourquoi, comme vous le comprendrez dans la suite de ce compte-rendu, nous ne les avons pas inscrites dans l'interface utilisateur à disposition. ### IV.a) Interface utilisateur:
<p style="text-align: justify;"> 
    Le fichier "Fenetre_paremetre" est <b> une interface graphique</b> réalisée via la bibliothèque pyhon tkinter. Elle autorise l'utilisateur à définir à la fois la situation initiale de l'écoulement granulaire(nombre de grains, position des paroi, etc.) de même que les paramètre de la simulation <a href="#ref_relation"> précédemment discuté </a>. Voici les différent paramètres qu'il est possible de définir à travers cette interface:
<ul>
    <li> Coeff. dir.: coefficient directeur des deux droites modélisant les parois du silo. </li>
    <li> Ord. origine: l'ordre à l'origine des deux droites. </li>
    <li> Debut Trou: ordonnée $y=cste$ à partir de laquelle les grains tombent.</li>
    <li> Durée simulation: le durée de la simulation souhaité. </li>
    <li> Hauteur bac: ordonnée $y=cste$ de la droite modélisant le bac de réception des grains. </li>
    <li> Hauteur grain: ordonnée du premier grain à placer dans le silo(le placement des autres grains se fait autmatiquement). </li>
    <li> Nombre de grains: le nombre de grains de la simulation. </li>
</ul>
L'interface propose aussi deux "checkbox" permettant de faire un choix binaire(oui/non, vrai/faux, 1/0):
<ul>
    <li> Afficher les grains: permet d'afficher dans l'interface le placement intial des grains. </li>
    <li> Activer la physique du bac: permet d'activer l'intéractions avec le bac, si cette case est coché le bac se comportera comme une paroi, sinon les grains arrivant dessus ne seront plus mise à jour et il deviendront totalement inerte, ces grains ne bougent plus et n'intéragissent plus avec leur environnement. </li>
</ul> 
</p>

<p style="text-align: justify;">
    <b> Lors du démarrage de notre simulateur </b> cette interface apparaît. L'utilisateur définit les paramètres souhaités puis appuie sur le bouton en bas de la fenêtre "Enregistrer". De cette manière, <b>les paramètres sont enregistrés sous la forme d'un fichier csv</b>(parametres.csv) et sont utilisés comme paramètres initiaux lors du lancement suivant, sinon ils ne le sont pas. La simulation se lance après la fermeture de l'interface avec les derniers paramètres inscrit à travers l'interface.
</p>

#### IV.b.1) Définition :
<p style="text-align: justify;">

<p style="text-align: justify;">
    Au sein de la partie "DEFINITION" le programme réalise toutes les définitions de variables, de tableaux de données ainsi que de script peremttant d'initialiser la simulation. A l'intérieur de cette sous-partie, nous définissions tout ce qui touche à/aux:

    Commençons par les définitions sur les variables liées aux grains. Les variable de rayon et de masse sont en réalité des tableaux. En effet, comme nous l'avons <a href="#variation_rayon">précisé précédemment</a> le rayon des grains varie de plus moins 20% par rapport à une valeur moyenne choisi. Le choix du rayon au sein de cette intervalle est totalement aléatoire. C'est pourquoi, chaque grain possède un rayon qui lui est propre. Pour la masse des grains elle est calculée grâce à la masse volumique inscrite dans notre programme. Ces deux variables sont donc des tableaux qui contiennent pour un grain d'indice $i$ les valeurs de rayon et de masse qui lui sont propres.
